```python
# Objective: Build a simple linear regression model to predict a numerical value based on a single feature.
```

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
from io import StringIO
```

```python
import os
print(os.getcwd())
```

```python
df = pd.read_csv ('winequality-red-data.csv')
df
```

```python
df.head()
```

```python
X = df['fixed acidity'] #Feature
Y = df['free sulfur dioxide'] #Target
```

```python
print(X.head())
print(Y.head())
```

```python
X = df['fixed acidity'].values.reshape(-1,1)
Y = df['free sulfur dioxide'].values
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression 
from sklearn.metrics import mean_squared_error, r2_score
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.2)
```

```python
model = LinearRegression()
model.fit(X_train, Y_train)
```

```python
prediction = model.predict(X_test) 
```

```python
# Evaluate the model using matrics like mean-squared-error
```

```python
msg = mean_squared_error(Y_test, prediction)
msg2 = r2_score(Y_test, prediction)
print(f"mean squared error : {msg:.2f}")
print(f"r2 score error : {msg2:.2f}")
```

```python
plt.scatter(X,Y,color = "Blue")
plt.plot(X, model.predict(X), color = "red")
plt.title("Linear Regression Model")
plt.xlabel("Feature")
plt.ylabel("Target")
plt.show()
```
